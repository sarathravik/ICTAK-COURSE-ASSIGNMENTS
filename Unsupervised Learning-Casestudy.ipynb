#  Trying out different clustering models in the wine dataset and finding the optimum number of clusters in each model and creating the model with the optimum number of clusters.

## 1.K-Means Clustering

```python
from sklearn.cluster import KMeans
#print within cluster sum of squares
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)
    
```

```python
#elbow method
plt.figure(figsize=(10,10))
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

From the above graph, it is clear that the optimum number of clusters is 3.

```python
#k-means with number  of clusters=3
kmeans=KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans=kmeans.fit_predict(data)
type(data)
y_kmeans
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data[:,0],data[:,1],c=y_kmeans)
plt.show()
```

## 2.Hierarchical Clustering

```python
#dendrogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(data,method='ward'))
plt.hlines(10,0,5000,colors='r',linestyle='--')
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Euclidean Distance')
plt.show()
```

Optimal number of Clusters=5

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage="ward")
aghc=cluster.fit_predict(data)
aghc
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data[:,0],data[:,1],c=aghc)
plt.show()
```

## 3.PCA

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(data)
x_pca=pca.transform(data)
x_pca.shape
```

```python
#plotting pca components
plt.figure(figsize=(10,10))
plt.scatter(x_pca[:,0],x_pca[:,1],cmap='plasma')
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
plt.show()
```
