```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=loan_df.drop(['loan_id','origination_date','first_payment_date','insurance_percent','insurance_type','m13'],axis=1)
Y=pd.DataFrame(loan_df['m13'])
```

```python
Y.shape
```

```python
#data preprocessing
loan_df.info()
```

```python
loan_df['financial_institution'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
a=['financial_institution']
for i in np.arange(len(a)):
    X[a[i]]=label_en.fit_transform(X[a[i]])
X.head()
```

```python
loan_df['loan_purpose'].value_counts()
```

```python
loan_df['source'].value_counts()
```

```python
#one hot encoding
#X=pd.get_dummies(X)
#X.head()
```

```python
#one hot encoding
source_en=pd.get_dummies(X['source'])
loanp_en=pd.get_dummies(loan_df['loan_purpose'])
```

```python
X=pd.concat([X,source_en,loanp_en],axis=1)
X.drop(['source','loan_purpose'],axis=1,inplace=True)
X.head()
```

```python
#feature engineering - adding new feature into the calculation from existing calculation
```

```python
X['credit_score']=X['co-borrower_credit_score']+X['borrower_credit_score']
X.drop(['borrower_credit_score','co-borrower_credit_score'],axis=1,inplace=True)
```

```python
X['mean']=X[['m1','m2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10', 'm11', 'm12']].mean(axis=1)
X['skew']=X[['m1','m2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10', 'm11', 'm12']].skew(axis=1)
X['kurtos']=X[['m1','m2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10', 'm11', 'm12']].kurt(axis=1)
X['sum']=X[['m1','m2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10', 'm11', 'm12']].sum(axis=1)
       
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.3,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#randomforest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import f1_score, confusion_matrix
```

```python
f1_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#feature importance
pd.Series(rf.feature_importances_,index=X.columns).sort_values(ascending=False)*100
```
